```python
file_bank=open('Bank_History.txt')
strr_bank=file_bank.readlines()

file_interval=open('Larger_Interval_History.txt')
strr_interval=file_interval.readlines()

file_no=open('No_Bank_History.txt')
strr_no=file_no.readlines()

acc_bank=[]
nmi_bank=[]
ari_bank=[]
f_bank=[]

acc_interval=[]
nmi_interval=[]
ari_interval=[]
f_interval=[]

acc_no=[]
nmi_no=[]
ari_no=[]
f_no=[]

loss_bank=[]
loss_interval=[]
loss_no=[]

for i in strr_bank:
    if 'Test NMI' in i:
        strs=i.split(' ')
        acc_bank.append(strs[12])
        ari_bank.append(strs[6])
        nmi_bank.append(strs[3])
        f_bank.append(strs[9])
    if 'Epoch' in i:
        strs=i.split(' ')
        loss_bank.append(strs[3])

for i in strr_interval:
    if 'Test NMI' in i:
        strs=i.split(' ')
        acc_interval.append(strs[12])
        ari_interval.append(strs[6])
        nmi_interval.append(strs[3])
        f_interval.append(strs[9])
    if 'Epoch' in i:
        strs=i.split(' ')
        loss_interval.append(strs[3])

for i in strr_no:
    if 'Test NMI' in i:
        strs=i.split(' ')
        acc_no.append(strs[12])
        ari_no.append(strs[6])
        nmi_no.append(strs[3])
        f_no.append(strs[9])
    if 'Epoch' in i:
        strs=i.split(' ')
        loss_no.append(strs[3])

```

```python
from matplotlib import pyplot as plt
import numpy as np

acc_bank=np.array(acc_bank,dtype=np.float32)
acc_interval=np.array(acc_interval,dtype=np.float32)
acc_no=np.array(acc_no,dtype=np.float32)

plt.figure(figsize=(12, 6))
plt.plot(np.log10(acc_bank[2:200]),label='Interval - 20')
plt.plot(np.log10(acc_interval[2:200]),label='Interval - 50')
plt.plot(np.log10(acc_no[2:200]),label='No Bank')
plt.tight_layout()
plt.legend()
plt.title('ACC Comparison in log form')
plt.show()
```

```python
ari_bank=np.array(ari_bank,dtype=np.float32)
ari_interval=np.array(ari_interval,dtype=np.float32)
ari_no=np.array(ari_no,dtype=np.float32)

plt.figure(figsize=(12, 6))
plt.plot(np.log10(ari_bank[2:200]),label='Interval - 20')
plt.plot(np.log10(ari_interval[2:200]),label='Interval - 50')
plt.plot(np.log10(ari_no[2:200]),label='No Bank')
plt.tight_layout()
plt.legend()
plt.title('ARI Comparison in log form')
plt.show()
```
